```python
import keras
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
model = keras.models.load_model('./TrainedModels/generator_model_200.h5', compile= False)
model2 = keras.models.load_model('./TrainedModels/generator_model_200_2.h5', compile= False)
```

```python
def plotter(model, n_samples=49):
    pnts = []
    plt.figure(figsize=(18,9))
    for i in range(n_samples):
        ran = np.random.randn(n_samples, 100)
        sqrt = int(np.sqrt(n_samples))
        plt.subplot(sqrt, sqrt, i+1)
        X = model.predict(ran)
        pnts.append(ran)
        X = (X + 1) / 2.0
        plt.imshow(X[0])
        plt.axis('off')
    plt.show()
    return pnts
```

```python
pnts = plotter(model, n_samples= 9)
```

```python
pnts = np.array(pnts)
p0 = np.array(pnts[1])
p1 = np.array(pnts[8])
```

```python
X = model.predict(p0)
X = (X + 1) / 2.0
plt.imshow(X[0])
plt.show()
X = model.predict(p1)
X = (X + 1) / 2.0
plt.imshow(X[0])
plt.show()
```

```python
def create_movie(model, p0, p1, n= 5):
    interp = np.linspace(p0, p1, num= n)
    for i in range(interp.shape[0]):
        x_inp = interp[i]
        X = model.predict(x_inp)[0]
        
        X = np.uint8(255 * X)
        
        video_name = 'Video.avi'
        height, width, layers = X.shape
        video = cv2.VideoWriter(video_name, 0, 1, (width,height))
        video.write(X)

        cv2.destroyAllWindows()
        video.release()
```

```python
create_movie(model= model, p0= p0, p1= p1)
```
