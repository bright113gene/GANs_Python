```python
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
def define_generator(latent_dim):
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(units= 256 * 4 * 4, input_dim= latent_dim))
    model.add(keras.layers.LeakyReLU(0.2))
    model.add(keras.layers.Reshape((4, 4, 256)))
    # 8 * 8 now
    model.add(keras.layers.Conv2DTranspose(filters= 128,
                                           kernel_size= (4,4),
                                           padding= 'same',
                                           strides= (2,2)))
    model.add(keras.layers.LeakyReLU(0.2))
    # 16 * 16 now
    model.add(keras.layers.Conv2DTranspose(filters= 128,
                                           kernel_size= (4,4),
                                           padding= 'same',
                                           strides= (2,2)))
    model.add(keras.layers.LeakyReLU(0.2))
    # 32 * 32 now
    model.add(keras.layers.Conv2DTranspose(filters= 128,
                                           kernel_size= (4,4),
                                           padding= 'same',
                                           strides= (2,2)))
    model.add(keras.layers.LeakyReLU(0.2))
    model.add(keras.layers.Conv2D(filters= 3,
                                  kernel_size= (3,3),
                                  padding= 'same',
                                  activation= 'tanh'))
    
    return model
```

```python
def generate_latent_points(latent_dim, n_samples):
    X = np.random.randn(latent_dim * n_samples)
    X = X.reshape((n_samples, latent_dim))
    return X
```

```python
def generate_fake_sample(g_model, latent_dim, n_samples):
    X = generate_latent_points(latent_dim= latent_dim,
                                  n_samples= n_samples)
    X = g_model.predict(X)
    y = np.zeros((n_samples, 1))
    return X, y
```

```python
latent_dim = 100
model = define_generator(latent_dim= latent_dim)
X_fake, y_fake = generate_fake_sample(g_model= model, latent_dim= latent_dim, n_samples= 9)
X_fake = (X_fake + 1.0 ) / 2.0
X_fake.shape, np.max(X_fake), np.min(X_fake)
```

```python
for i, im in enumerate(X_fake):
    plt.subplot(3,3,i+1)
    plt.imshow(im)
plt.show()
```
